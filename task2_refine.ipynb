````python
def check_different_predictions(result_data):
    """
    Check and return items where 'predict' is in valid answers and differs from 'gold'.
    Input format: list of dicts with 'question_id', 'predict', and 'gold' keys
    Returns list of items where predictions differ from gold answers and 'predict' is valid
    """
    different_items = []
    valid_answers = {"sai", "đúng", "a", "b", "c", "d"}

    for item in result_data:
        if item["predict"].lower() in valid_answers:
            different_items.append(item)

    return different_items


different_items = check_different_predictions(data["results"])
len(different_items)
````

````python
import os
from openai import OpenAI


# model_name = "DeepSeek-R1-Distill-Llama-8B"
# model_name = "Qwen2.5-7B-Instruct"
model_name = "Llama-3.1-8B-Instruct"
from dotenv import load_dotenv

load_dotenv("llm-fpt/.env")


class FPTLlm:
    def __init__(self, model_name: str = "Qwen2.5-7B-Instruct"):
        self.client = OpenAI(
            api_key=os.getenv("FPT_API_KEY"), base_url="https://mkp-api.fptcloud.com"
        )
        self.model_name = model_name

    def complete(self, input: str) -> str:
        response = self.client.chat.completions.create(
            model=self.model_name,
            messages=[{"role": "user", "content": input}],
            max_tokens=1000,
            temperature=0.01,
        )
        return response.choices[0].message.content.strip()


llm = FPTLlm(model_name=model_name)
````

````python
def extract_answer(text):
    import re

    match = re.search(r"```answer\s*([\s\S]+?)\s*```", text)
    answer = match.group(1).strip() if match else text.strip()
    return answer.replace(".", "")
````

````python
from tqdm import tqdm

for item in tqdm(different_items_update):
    refine_predict = llm.complete(
        REFINE_PROMPT.format(
            question=item["question"],
            context=item["context"],
            predict=item["predict"],
        )
    )
    item["refine_predict"] = extract_answer(refine_predict)
````

````python
# Evauluaton LLama
def compute_string_match_accuracy(inputs):
    """
    Compute accuracy based on exact string matching between refined predictions and gold labels.
    Args:
        inputs: list of dicts, each containing 'refine_predict' and 'gold' keys
    Returns:
        float: accuracy score
    """
    if not inputs:
        return 0.0

    correct = sum(
        1
        for item in inputs
        if item["refine_predict"].strip().lower()
        == item["gold"].strip().lower()
    )
    return correct / len(inputs)


acc = compute_string_match_accuracy(different_items_update)
acc
````
